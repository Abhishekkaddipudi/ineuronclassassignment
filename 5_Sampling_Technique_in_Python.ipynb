 <h3>Population</h3>
 
- The population is the set of all observations (individuals, objects, events, or procedures) and is usually very large and diverse.

<h3>Sample</h3>

- A sample is a subset of observations from the population that ideally is a true representation of the population.

<div class="alert alert-block alert-info">
    <h3 style="margin:0"><b>Loading an example dataset from seaborn</b></h3>
</div>

```python
import seaborn as sns
df=sns.load_dataset('titanic')
```

<div class="alert alert-block alert-info">
<h3 style="margin:0" > <b>1. Random Sampling from numpy library</b></h3>
</div>

```python
import numpy as np
random=np.random.choice(df.index,replace=False,size=10)
df.iloc[list(random)]
#randomly selecting index value
#choice method replace argument is defining with or without repetation
#size argument return number of sample
```

<div style="height:100px">
    <p></p>

</div>

<div class="alert alert-block alert-info">
<h3 style="margin:0" > <b>Random Sampling from random module</b></h3>
</div>

```python
import random
select_index=[]
for i in range(1,10):
    select_index.append(random.randint(1,len(df)))
    #generating random index 
df.iloc[select_index]
```

    <h3 style="margin:0"> 2. Systemic sampling :targeting population from head or tail position or from certain fixed intervals </h3>
 </div>

```python
df.head(5)
```

```python
df.tail(5)
```

```python
select_index=list(range(1,len(df),50))
df.iloc[select_index]
#we choose from 50 fixed interval
```

<h3 style="margin:0">3. By dividing the population into multiple groups based on certain criteria.</h3>
</div>

```python
import pandas as pd
data=pd.DataFrame()
for i in df["embark_town"].unique():
    data=pd.concat([data,df[df["embark_town"]==i].head(3)])
display(data)
#diving each by embarked town and concating first 3 entries
```
